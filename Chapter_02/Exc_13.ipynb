```python
import pandas as pd
import numpy as np
from numpy import nan
```

```python
temp_data = pd.read_csv('../data_sets/nyc-temps.txt').squeeze()
temp_data
```

```python
df = pd.DataFrame({'Temperature':temp_data, 'Hour': [0, 3, 6, 9, 12, 15, 18, 21]*91})
df
```

```python
df.describe()
```

```python
df.loc[df['Hour'].isin([3, 6]), 'Temperature'] = nan
```

```python
df.interpolate().describe()
```

# Beyond 1
### How does the behavior of interpolate change if you use method='nearest'?

```python
df.interpolate(method='nearest').describe()
```

### Apparently methodf='nearest' does not change behaviour of interpolation

# Beyond 2
### Let’s assume the equipment works fine around the clock but fails to record readings at –1 degrees and below. Are the interpolated values similar to the real (missing) values they replace? Why or why not?

```python
df.loc[df['Temperature'] <= -1, 'Temperature'] = nan
df.interpolate().describe()
```

### No they are not similar to the real values because we removed a chunk of array thus shifting mean etc.

# Beyond 3
### A cheap solution to interpolation is to replace NaN values with the column’s mean. Do this (with the missing values from –1 and below), and compare the new mean and median. Again, why are (or aren’t) these values similar to the original ones?

```python
df = pd.DataFrame({'Temperature':temp_data, 'Hour': [0, 3, 6, 9, 12, 15, 18, 21]*91})
df
```

```python
df.loc[df['Temperature'] <= -1, 'Temperature'] = df['Temperature'].mean()
df.interpolate().describe()
```

### Mean shifted to lower values. Median unchanged.
