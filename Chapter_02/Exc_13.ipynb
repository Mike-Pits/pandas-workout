```python
import pandas as pd
import numpy as np
from numpy import nan
```

```python
temp_data = pd.read_csv('../data_sets/nyc-temps.txt').squeeze()
temp_data
```

```python
df = pd.DataFrame({'Temperature':temp_data, 'Hour': [0, 3, 6, 9, 12, 15, 18, 21]*91})
df
```

```python
df.describe()
```

```python
df.loc[df['Hour'].isin([3, 6]), 'Temperature'] = nan
```

```python
df.interpolate().describe()
```

# Beyond 3
### A cheap solution to interpolation is to replace NaN values with the columnâ€™s mean. Do this (with the missing values from â€“1 and below), and compare the new mean and median. Again, why are (or arenâ€™t) these values similar to the original ones?

```python
df = pd.DataFrame({'Temperature':temp_data, 'Hour': [0, 3, 6, 9, 12, 15, 18, 21]*91})
df
```

```python
df.loc[df['Temperature'] <= -1, 'Temperature'] = df['Temperature'].mean()
df.interpolate().describe()
```

### Mean shifted to lower values. Median unchanged.
