### Create a data frame from the file nyc-parking-violations-2020.csv. We are only interested in a handful of the columns:

### – Plate ID
### – Registration State
### – Issue Date (a string in MM/DD/YYYY format, always followed by 12:00:00AM)
### – Vehicle Make
### – Street Name
### – Vehicle Color

```python
df = pd.read_csv('../data_sets/nyc-parking-violations-2020.csv', parse_dates=True,
                 usecols=['Date First Observed', 'Plate ID', 'Registration State',
                          'Issue Date', 'Vehicle Make', 'Street Name', 'Vehicle Color'])
df.head()
```

### Set the data frame’s index to the Issue Date column.

```python
df = df.set_index('Issue Date')
```
