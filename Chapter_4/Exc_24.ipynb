```python
import pandas as pd
```

```python
df = pd.read_csv('/home/mike-pi/Documents/coding/Excs/pandas-workout/pandas-workout/data_sets/olympic_athlete_events.csv',
                 usecols=['Age', 'Height', 'Team', 'Year', 'Season', 'Sport', 'Medal']
    )

df.head()
```

```python
df = df.loc[df['Team'].isin(['United States', 'Great Britain', 'France', 'Switzerland', 'China', 'India'])]

df.head()
```

```python
df = df.loc[df['Year'] >= 1980]
```

### What was the average age of Olympic athletes? In which country do players appear to consistently be the youngest?

```python
df.pivot_table(index='Year', columns='Team', values='Age')
```

### How tall were the tallest athletes in each sport in each year?

```python
df.pivot_table(index='Year', columns='Sport', values='Height', aggfunc='max')
```

### How many medals did each country earn each year?

```python
df.pivot_table(index='Year', columns='Team', values='Medal', aggfunc='count', fill_value=0)
```

# Beyond 1
### Create a pivot table that shows the number of medals each team won per year, with the index including the year and the season in which the games took place.

```python
df.pivot_table(index=['Year', 'Season'], columns='Team', values='Medal', aggfunc='count', fill_value=0)
```

# Beyond 2
### Create a pivot table that shows both the average age and the average height per year per team.

```python
df.pivot_table(index='Year', 
               columns='Team', 
               values=['Age', 'Height'], 
               aggfunc='mean')
```

# Beyond 3
### Create a pivot table that shows the average age and the average height per year, per team, broken up by year and season.

```python
df.pivot_table(index=['Year'], 
               columns='Team', 
               values=['Age', 'Height'], 
               aggfunc='max') 
```
